```python
import os
from PIL import Image

def load_data(img_dir, text_dir):
    data_pairs = []
    for img in os.listdir(img_dir):
        img_path = os.path.join(img_dir, img)
        text_path = os.path.join(text_dir, img.replace(".png", ".md"))
        if os.path.exists(text_path):
            with open(text_path, 'r') as f:
                text = f.read().strip()
            data_pairs.append((img_path, text))
    return data_pairs
data_pairs = load_data('img/rain/', 'text/rain/')
```

```python
# TODO: train CLIP
inputs = processor(
    text=[pair[1] for pair in data_pairs],
    images=[Image.open(pair[0]) for pair in data_pairs],
    return_tensors="pt",
    padding=True,
    truncation=True,
    max_length=77,
    )
```
